```python
from matplotlib import pyplot as plt
import pandas as pd
from scipy.signal import spectrogram
```

```python
import numpy as np


df = pd.DataFrame({"val": [np.sin(2 * np.pi * t) for t in range(100)]})
df
```

```python
f, t, Sxx = spectrogram(
    x=df["val"],
    fs=10,
    nperseg=30,
    mode="magnitude",
)

plt.figure()
plt.pcolormesh(t, f, Sxx, shading="gouraud")
plt.title(f"Spectrogram for val")
plt.ylabel("Frequency [Hz]")
plt.xlabel("Time [sec]")
plt.colorbar(label="Magnitude")
plt.show()
```
